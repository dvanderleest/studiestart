## Assignment 3: Protein
The lecture discussed how proteins are translated from DNA. Of course, this all happens neatly in our cells. If we know the DNA sequence, we also know the protein sequence. Of course we can also program that out.

In assignment 1 we discussed what a function is and what it looks like. The teacher has written a function for you below that will perform the translation for you, but you still have to call it yourself.

```python
# This is the function. You don't have to change anything in this code, but you do have to execute this field.
# Otherwise, Python doesn't know what the function definition is.
def translate(dna_sequence):
    
    dna_sequence = dna_sequence.lower()
    
    start_codon = 'atg'
    stop_codons = ['taa','tag','tga']
    
    code = {'ttt': 'F', 'tct': 'S', 'tat': 'Y', 'tgt': 'C',
            'ttc': 'F', 'tcc': 'S', 'tac': 'Y', 'tgc': 'C',
            'tta': 'L', 'tca': 'S', 'taa': '*', 'tga': '*',
            'ttg': 'L', 'tcg': 'S', 'tag': '*', 'tgg': 'W',
            'ctt': 'L', 'cct': 'P', 'cat': 'H', 'cgt': 'R',
            'ctc': 'L', 'ccc': 'P', 'cac': 'H', 'cgc': 'R',
            'cta': 'L', 'cca': 'P', 'caa': 'Q', 'cga': 'R',
            'ctg': 'L', 'ccg': 'P', 'cag': 'Q', 'cgg': 'R',
            'att': 'I', 'act': 'T', 'aat': 'N', 'agt': 'S',
            'atc': 'I', 'acc': 'T', 'aac': 'N', 'agc': 'S',
            'ata': 'I', 'aca': 'T', 'aaa': 'K', 'aga': 'R',
            'atg': 'M', 'acg': 'T', 'aag': 'K', 'agg': 'R', 
            'gtt': 'V', 'gct': 'A', 'gat': 'D', 'ggt': 'G',
            'gtc': 'V', 'gcc': 'A', 'gac': 'D', 'ggc': 'G',
            'gta': 'V', 'gca': 'A', 'gaa': 'E', 'gga': 'G',
            'gtg': 'V', 'gcg': 'A', 'gag': 'E', 'ggg': 'G'}
    
    start_position = dna_sequence.find(start_codon)
    dna_sequence = dna_sequence[start_position:]
    protein_sequence = ''
    
    for i in range(0, len(dna_sequence), 3):
        codon = dna_sequence[i:i + 3]
        if codon not in stop_codons:
            protein_sequence = protein_sequence + code[codon]
        else:
            return protein_sequence
```
